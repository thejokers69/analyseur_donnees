```python
# Étape 1 : Importer les bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Étape 2 : Générer un DataFrame avec des colonnes numériques
np.random.seed(42)  # Pour la reproductibilité
data = {
    "ID": np.arange(1, 1001),  # Identifiants de 1 à 1000
    "Age": np.random.randint(18, 65, size=1000),  # Âges aléatoires entre 18 et 65
    "Salary": np.random.randint(30000, 120000, size=1000),  # Salaires aléatoires
    "Height": np.random.uniform(1.5, 2.0, size=1000),  # Taille en mètres
    "Weight": np.random.uniform(50, 120, size=1000),  # Poids en kg
}

df = pd.DataFrame(data)

# Étape 3 : Sauvegarder le DataFrame en formats CSV et Excel
csv_file_name = "large_numeric_data.csv"
excel_file_name = "large_numeric_data.xlsx"

df.to_csv(csv_file_name, index=False)
df.to_excel(excel_file_name, index=False)

print(f"Fichiers créés :\n- {csv_file_name}\n- {excel_file_name}")

# Étape 4 : Charger les fichiers
df_csv = pd.read_csv(csv_file_name)
df_excel = pd.read_excel(excel_file_name)

# Étape 5 : Identifier les colonnes numériques
numeric_columns_csv = df_csv.select_dtypes(include="number").columns.tolist()
numeric_columns_excel = df_excel.select_dtypes(include="number").columns.tolist()

print("Colonnes numériques dans le fichier CSV :", numeric_columns_csv)
print("Colonnes numériques dans le fichier Excel :", numeric_columns_excel)

# Étape 6 : Générer des visualisations
# Exemple : Histogramme pour la colonne 'Salary'
if "Salary" in numeric_columns_csv:
    plt.figure(figsize=(10, 6))
    sns.histplot(df_csv["Salary"], kde=True, color="blue")
    plt.title("Histogram of Salary")
    plt.xlabel("Salary")
    plt.ylabel("Frequency")
    plt.show()
else:
    print(
        "La colonne 'Salary' n'est pas disponible dans le fichier CSV pour un histogramme."
    )

# Exemple : Scatter plot pour 'Age' et 'Salary'
if "Age" in numeric_columns_csv and "Salary" in numeric_columns_csv:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="Age", y="Salary", data=df_csv, color="green")
    plt.title("Scatter Plot: Age vs Salary")
    plt.xlabel("Age")
    plt.ylabel("Salary")
    plt.show()
else:
    print(
        "Les colonnes 'Age' et 'Salary' ne sont pas disponibles pour un scatter plot."
    )
```
