```python
import pandas as pd

# Création d'un DataFrame avec des colonnes numériques
data = {
    "ID": [1, 2, 3, 4, 5],
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000],
    "Height": [5.5, 6.0, 5.8, 5.9, 6.1],
}

df = pd.DataFrame(data)

# Sauvegarder le fichier au format CSV
file_name = "numeric_data.csv"
df.to_csv(file_name, index=False)

print(f"Fichier {file_name} créé avec succès.")
```

```python
# Charger le fichier CSV
df = pd.read_csv(file_name)

# Sélectionner uniquement les colonnes numériques
numeric_columns = df.select_dtypes(include="number").columns.tolist()

print("Colonnes numériques détectées :", numeric_columns)

# Vérifier si des colonnes numériques sont présentes
if not numeric_columns:
    print("Aucune colonne numérique disponible pour la visualisation.")
else:
    print("Les colonnes numériques sont prêtes pour la visualisation.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Exemple : Générer un histogramme pour la colonne 'Salary'
if "Salary" in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df["Salary"], kde=True)
    plt.title("Histogram of Salary")
    plt.xlabel("Salary")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("La colonne 'Salary' n'est pas numérique.")
```

```python
# Sauvegarder le DataFrame en format Excel
excel_file_name = "numeric_data.xlsx"
df.to_excel(excel_file_name, index=False)

print(f"Fichier {excel_file_name} créé avec succès.")
```
